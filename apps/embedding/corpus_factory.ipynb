```python
from collections import Counter

tokens = [1,2,3,1]
freqs1 = sorted(Counter(tokens).items(), key=lambda x: x[1], reverse=True)
print(freqs1)

def f1(tokens):
    for x in tokens:
        yield x
    
g1 = f1(tokens)
freqs2 = sorted(Counter(g1).items(), key=lambda x: x[1], reverse=True)
print(freqs2, type(g1))
```

```python
from sgd_nlp.common.read_file import yield_tokens_from_docs
from collections import Counter

def _init_documents(docs_dir):
    # 递归枚举所有子文件路径
    def recursive_list_files(root_dir):
        f_list = []
        for root, dirs, files in os.walk(root_dir, topdown=False):
            f_list += ([os.path.join(root, fi) for fi in files])
            if dirs:  # 非空
                for di in dirs:
                    f_list += (recursive_list_files(os.path.join(root, di)))
        return f_list

    if os.path.isfile(docs_dir):  # case1： 单个文件
        return [docs_dir]
    else: # case2：递归遍历文件夹
        return recursive_list_files(docs_dir)
    
docs = _init_documents(corpus_dir)
print(docs)
        
tokens = yield_tokens_from_docs(docs)
print(tokens)
# print("[Debug]", [x for x in tokens][:5], type(tokens))

_token_freqs = sorted(Counter(tokens).items(), key=lambda x: x[1], reverse=True)
print(_token_freqs)
```
